**Calculating Wet Soil Skewness**

```python
import glob
from scipy.stats import skew

folders = glob.glob('/Users/braydennoh/Desktop/segment/moisture/test1/wet')

imagenames_list = []
for folder in folders:
   for f in glob.glob(folder+'/*.jpg'):
       imagenames_list.append(f)

read_images = []        
result_arr = []

for image in imagenames_list:
    img = cv2.imread(image)
    img_Luv = cv2.cvtColor(img, cv2.COLOR_RGB2Luv)
    gray = cv2.cvtColor(img_Luv, cv2.COLOR_BGR2GRAY)
    result_arr.append(gray)


result_arr = np.concatenate(result_arr)

plt.hist(result_arr)
from scipy.stats import skew
  
# Calculate the skewness
skew = np.mean(skew(result_arr, axis=0, bias=True))

print(skew)
```
